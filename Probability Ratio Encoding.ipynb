```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('../CSV Files/titanic_train.csv', usecols=['cabin', 'survived'])
```

```python
df.head()
```

```python
df.cabin.fillna('Missing', inplace=True)
```

```python
df.head()
```

```python
df.dtypes
```

```python
df.cabin.unique()
```

```python
df['cabin'] = df.cabin.str[0]
df.head()
```

```python
df.cabin.unique()
```

```python
prob_df = df.groupby(['cabin'])['survived'].mean()
```

```python
df.groupby(['cabin', 'survived']).size().reset_index().rename(columns={0: 'Count'})
```

```python
prob_df
```

```python
prob_df = pd.DataFrame(prob_df)
```

```python
prob_df
```

```python
prob_df['Died'] = 1-prob_df['survived']
prob_df.head()
```

```python
prob_df['Probability_Ratio'] = prob_df['survived'] / prob_df['Died']
```

```python
prob_df
```

```python
prob_df.rename(columns={'survived': 'Survived'}, inplace=True)
```

```python
prob_df
```

```python
prob_ratio = prob_df['Probability_Ratio'].to_dict()
```

```python
df['cabin'] = df['cabin'].map(prob_ratio)
df.head()
```
